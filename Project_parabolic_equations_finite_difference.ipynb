## Parabolic equations

### General information and classification 


Parabolic partial differential equations are used to describe variety of time dependent phenomena. such as heat conduction or pricing of derivative investment instruments. To classify PDE (Partial Differential Equation) as parabolic we must first define the general form of second order, linear, constant-coefficient PDE whose solution is some function $u(x, t)$, where $x$ and $t$ are independent real variables. Such equation takes form: $\newline$

$$Au_{xx} + Bu_{xt} + Cu_{tt} + Du_{x} + Eu_{t} + Fu = G. $$

The classification of this equation depends on the coefficients, exactly speaking the sign of $B^2 - 4AC$. Based on that we have three possibilities of the equation. Hence, if: $\newline$
- $B^2 - 4AC < 0$, then the equation is elliptic,
- $B^2 - 4AC = 0$, then the equation is parabolic,
- $B^2 - 4AC > 0$, then the equation is hyperbolic.

The examples of parabolic PDEs can be Diffusion-Convection equation and Black-Scholes equation which we are going to analyse.

## Finite Difference Method

The Finite Difference Method is a numerical technique used to solve partial differential equations (PDEs) by approximating the derivatives of the unknown function using finite difference approximations. This method converts the continuous PDE problem into a discrete problem that can be solved using algebraic equations.

### Basic Idea

The basic idea behind the Finite Difference Method is to replace the derivatives in the PDE with finite difference approximations. This is done by dividing the spatial domain into a grid of discrete points and approximating the derivatives at these points using finite differences. The PDE is then transformed into a system of algebraic equations that can be solved using numerical methods.

### Discretization

To apply the Finite Difference Method, the spatial domain is discretized into a grid. The grid points are denoted by (i, j) or (x_i, y_j), where i and j represent the indices along the x and y axes, respectively. The grid spacing is represented by Δx and Δy, which are the distances between adjacent grid points in the x and y directions, respectively.

### Approximating Derivatives

The derivatives in the PDE are approximated using finite difference formulas. The central difference approximation is commonly used, which expresses the derivative at a point (i, j) as a linear combination of function values at neighboring grid points. 

Let's start from methods used for approximation first order derivative:

$\delta_+ y(x) := \frac{y(x+h) - y(x)}{h}$ - forward difference form 

$\delta_- y(x) := \frac{y(x) - y(x-h)}{h}$ - forward difference form 

$\delta_0 y(x) := \frac{1}{2} (\delta_- y(x) + \delta_+ y(x)) = \frac{y(x+h) - y(x-h)}{2h}, h >  0$ - centred difference





### Discretized PDE Equation

By substituting the finite difference approximations into the original PDE, we obtain a discretized equation. This equation relates the function values at the grid points and their neighboring points.

### Solving the System of Equations

The system of algebraic equations obtained from the discretized PDE can be solved using various numerical methods, such as direct methods (e.g., Gaussian elimination) or iterative methods (e.g., Jacobi method, Gauss-Seidel method). The choice of method depends on the specific problem and the desired accuracy and efficiency.

### Boundary Conditions

Boundary conditions are crucial in the Finite Difference Method. They are used to specify the values of the unknown function at the boundaries of the spatial domain. Different types of boundary conditions can be applied, such as Dirichlet boundary conditions (prescribing the function values) or Neumann boundary conditions (prescribing the derivatives). These conditions are incorporated into the system of equations to obtain a well-posed problem.

### Space discretization
To solve the parabolic partial differential equations using the explicit and implicit finite difference method, we divide the time interval [0, T] into $M$ equally sized subintervals of length $\Delta t$. Similarly we discretize the space variable.  


### Advantages and Limitations

The Finite Difference Method offers several advantages, including its simplicity and flexibility. It can handle complex geometries and various types of PDEs. However, it has some limitations. The accuracy of the method depends on the grid resolution, and it may require a fine grid to achieve high accuracy. The method may also suffer from stability issues for certain types of PDEs.


## Diffusion-Convection Equation 

In this part we will provide general form and solution of 1 dimensional Diffusion-Convection equation, which is a connection of both diffusion (also known as heat) and convection equations.

### Diffusion Equation

Diffusion equation is mathematical description of diffusion process, which involves dispersion of particles caused by random movements and collisions of those particles with each other. Such equation takes form: $\newline$

$$ \frac{∂u}{\partial t} = c^2 \frac{∂^2 u}{∂x^2}, \tag{1}\label{D eq} $$ $\newline$

where: $\newline$
- $c^2$ - diffusion coefficient.

In our case function $u$ can be interpreted as the temperature of the substance.

### Convection equation 

Convection process is a heat transfer, by movement of the fluid particles. In this process, rate of change of $u$ with respect to time is proportional to the rate of change of $u$ with respect to space variable. It can be expressed as: $\newline$

$$ \frac{∂u}{\partial t} = -b \frac{∂u}{∂x}, \tag{2}\label{C eq} $$ $\newline$

where: $\newline$
- $b$ - velocity of the fluid flow.

### Diffusion-Convection Equation

At this point we can define diffusion-convection equation as combination of $\eqref{D eq}$ and $\eqref{C eq}$ equations. It can be written as follows: $\newline$

$$\frac{∂u}{\partial t} = c^2 \frac{∂^2 u}{∂x^2} -b \frac{∂u}{∂x}, \tag{3}\label{D-C eq} $$

This equation describes transfer of heat or other physical quantities, in some system, where both diffusion and convection processes occurs. As there is only one space variable we consider one-dimensional systems.

### Solution of Diffusion-Convection Equation

To provide solution of diffusion-convection equation we must first specify domain, initial and boundary conditions.

#### Domain 

Transfer of mentioned physical quantities is conditioned by time ($t$) and space ($x$) variable. We consider systems where transfer takes place in only one dimension, length of such system can be marked as $L$. Therefore domain is specified as follows:

$$ x \in (0,L) \;\;\; \text{and} \;\;\ t > 0. $$

#### Initial Condition

Initial condition is the set of the known values of function $u(x, t)$ at the initial time (for example $t = 0$). In our case we will consider general form of such set given by some function $f(x)$. Therefore initial condition can be written as follows:

$$ u(x, 0) = f(x) $$

#### Boundary Conditions

Boundary conditions describe interaction between studied system and its surrounding. We will consider 3 types of boundary conditions.

- Dirichlet boundary conditions stands for prescribed values of function at each boundary. Other words we make assumption that value of function cannot change no metter what happens in the system. By seting such value as 0 we get homogeneous Dirichlet boundary conditions which can be written as follows: $\newline$

\begin{cases}
u(0,t) = 0 \\
u(L,t) = 0
\end{cases} $\newline$

- Neumann boundary conditions allow value of function to change at each boundary depending on what happens in system. Therefore homogeneous Neumann boundary conditions can be written as: $\newline$

\begin{cases}
\frac{∂}{∂x} u(0,t) = 0 \\
\frac{∂}{∂x} u(L,t) = 0
\end{cases} $\newline$

- Robin boundary conditions are based on two laws. First is Newton's law of cooling where heat transfer depends on difference between examined system and its surrounding. Second is Fourier's law, which tells us that direction of heat transfer is from warmer to cooler region. Mathematical description of those two laws combined together in its homogeneus version (where surrounding temperature is set to 0) can be written as follows: $\newline$

\begin{cases}
\frac{∂}{∂x} u(0,t) - C_1 u(0,t) = 0 \\
\frac{∂}{∂x} u(L,t) + C_2 u(L,t) = 0
\end{cases} $\newline$

$\;\;\;\;\;\;\;\;$ where $C_1$ and $C_2$ are some constants

#### Transforming Diffusion-Convection Equation into the Diffusion Equation 

To solve diffusion-convection equation we must first transform it into the pure diffusion equation as this allows us to use separation of variables method. From now we will also use subscript notation therefore equation $\eqref{D-C eq}$ can be rewritten as:

$$ u_{t} = c^2 u_{xx} - b u_{x}.  \tag{4}\label{D-C eq sub} $$

At this point we need to make an assumption that function $u(x,t)$ can be written as product of two functions both depending on $x$ and $t$ which can be marked as $y(x,t)$ and $z(x,t)$. Therefore we have:


$$ u(x,t) = z(x,t) y(x,t). \tag{5}\label{D-C eq product} $$


Now we can find function $z(x,t)$ such that $bu_{x}$ part, from equation $\eqref{D-C eq}$, disappears. By plugging $\eqref{D-C eq product}$ into $\eqref{D-C eq sub}$ we obtain:

$$ z_{t} y + z y_{t} = c^2 (z_{xx}y + 2z_{x}y_{x} + zy_{xx}) - b(z_{x}y + zy_{x}),$$ $\newline$

and after few transformations we get:

$$ y_{t} = c^2y_{xx} + \left(2c^2 \frac{z_{x}}{z} -b\right)y_{x} - \left(b 
\frac{z_{x}}{z} - c^2 \frac{z_{xx}}{z} + \frac{z_{t}}{z}\right)y.$$ $\newline$

Now we see that in order to obtain pure diffusion equation, following conditions must be satisfied: $\newline$

$$ 2c^2 \frac{z_{x}}{z} - b = 0 \tag{6}\label{ODE condition 1}, $$ $\newline$
$$ b \frac{z_{x}}{z} - c^2 \frac{z_{xx}}{z} + \frac{z_{t}}{z} = 0 \tag{7}\label{PDE condition 2},$$ $\newline$

which leads us to solving two differential equations. Starting with equation $\eqref{ODE condition 1}$, which is typical ordinary differential equation solved by integrating with respect to variable $x$. solution of such equation is: $\newline$

$$ z(x,t) = C(t) \exp{\left(\frac{b}{2c^2}x \right)}. \tag{8}\label{ODE 1 sol gen}$$ $\newline$

To find function $C(t)$ we will use fact that $z(x,t)$ must also satisfy equation $\eqref{PDE condition 2}$ which after multiplying by $z$ takes form: $\newline$

$$ z_{t} = c^2 z_{xx} - b z_{x}. \tag{9}\label{new PDE}$$ 

Now we can calculate partial derivatives based on equation $\eqref{ODE 1 sol gen}$. By doing so we obtain: $\newline$

$$ z_{t} = \frac{C_t}{C}z, $$ $\newline$

$$ z_{x} = \frac{b}{2c^2} z, $$ $\newline$
 
$$ z_{xx} = \frac{b^2}{4c^4} z, $$ $\newline$

and by plugging those derivatives into equation $\eqref{new PDE}$ we get: $\newline$

$$ \frac{C_t}{C}z = -\frac{b^2}{4c^2} z, $$

after dividing by $z$ we obtain ordinary differential equation in the form of: $\newline$

$$ \frac{C_t}{C} = -\frac{b^2}{4c^2}. \tag{10}\label{new ODE} $$ $\newline$

By solving equation $\eqref{new ODE}$ we obtain function $C(t)$ defined as: $\newline$

$$ C(t) = D \exp{\left( \frac{-b^2}{4c^2}t \right)}, \tag{11}\label{new ODE sol} $$ 

and by plugging $\eqref{new ODE sol}$ into $\eqref{D-C eq product}$ we get: $\newline$

$$ u(x,t) = D \exp{\left( \frac{-b^2}{4c^2}t \right)} \exp{\left(\frac{b}{2c^2}x \right)} y(x,t), \tag{12}\label{D-C transformed} $$ $\newline$

where $D$ is some constant not depending on either $x$ and $t$. $\newline$

Finally by substituting $\eqref{D-C transformed}$ into $\eqref{D-C eq sub}$ and simplifying we get diffusion equation: $\newline$

$$ y_{t} = c^2 y_{xx}. \tag{13}\label{heat eq} $$ $\newline$

Such a transformation also forces a change in the initial condition. Therefore we get:

$$ u(x,0) = z(x,0) y(x,0) = f(x), $$ $\newline$

$$ D\exp{\left(\frac{b}{2c^2}x \right)} y(x,0) = f(x), $$ $\newline$
$$ y(x, 0) = D\exp{\left( - \frac{b}{2c^2}x \right)} f(x). \tag{14}\label{new initial} $$

At this point as we work with pure diffusion equation we can use separation of variables method. In other words we can write down function $y(x,t)$ as product of two functions, each depending on one variable, so we get:

$$ y(x,t) = X(x)T(t). \tag{15}\label{separation} $$ $\newline$

By plugging equation $\eqref{separation}$ into equation $\eqref{heat eq}$ we get: $\newline$

$$ X(x)T'(t) = c^2 X''(x)T(t), $$ $\newline$

and using simplified notation we obtain:

$$ \frac{X''}{X} = \frac{1}{c^2}\frac{T'}{T} = Λ, \tag{16}\label{equal lambda} $$ $\newline$

where $Λ$ is some constant, as equality must be satisfied at any point $t$ and $x$. At this point we also need to use specific boundary conditions.

#### Solution for Dirichlet Boundary Conditions

For system described by diffusion-convection equation $\eqref{D-C eq sub}$ Dirichlet boundary conditions are: $\newline$

\begin{cases}
u(0,t) = 0 \\
u(L,t) = 0,
\end{cases} $\newline$

As transformed diffusion-convection equation $\eqref{D-C eq sub}$ into diffusion equation $\eqref{heat eq}$ we also have to find coresponding boundary conditions: $\newline$

from $u(0, t) = 0$ we get: $\newline$

$$ D \exp{\left( -\frac{b^2}{4c^2}t \right)}y(0,t) = 0 \implies y(0,t) = 0, $$ $\newline$

from $u(L,t) = 0$ we get: $\newline$

$$ D \exp{\left( -\frac{b^2}{4c^2}t \right)}\exp{\left( \frac{b}{2c^2}L \right)}y(L,t) = 0 \implies y(L,t) = 0. $$ $\newline$

We can see that in this case our boundary conditions remains the same, but for function $y$, so we have:

\begin{cases}
y(0,t) = 0 \\
y(L,t) = 0.
\end{cases} $\newline$

Now we can substitute new boundary conditions to our diffusion equation:

for $y(0,t) = 0$ we have:

$$ X(0)T(t) = 0 \implies X(0) = 0, \tag{17}\label{boundary X0} $$

for $y(L,t) = 0$ we have:

$$ X(L)T(t) = 0 \implies X(L) = 0. \tag{18}\label{boundary XL} $$ $\newline$

At this point we have to consider three cases of $Λ$ depending on its sign. Due to boundary conditions we will start solving our equation from function $X$. We will also use notations $Λ = λ^2$ and $Λ = -λ^2$ (depending considered on the case) in order for calculations to be more clear. $\newline$

$\underline{Case}$: $Λ = λ^2 = 0$

As we provide solution for $X$ equation $\eqref{equal lambda}$ takes form: $\newline$

$$ X'' = 0, $$ $\newline$

and general solution is: $\newline$

$$ X(x) = c_1 x + c_2, $$

now using conditions $\eqref{boundary X0}$ and $\eqref{boundary XL}$ we get 

$$ X(0) = c_2 = 0, $$

$$ X(L) = c_1 L = 0 \implies c_1 = 0. $$ $\newline$

Since both $c_1$ and $c_2$ are equal to zero case $λ^2$ is trivial. $\newline$

$\underline{Case}$: $Λ = λ^2 > 0$

In this case we consider equation:

$$ X'' = λ^2X $$ $\newline$

General solution of such ordinary differential equation is:

$$ X(x) = c_3 \exp{(λx)} + c_4 \exp{(-λx)},$$ $\newline$

now using conditions $\eqref{boundary X0}$ and $\eqref{boundary XL}$ we get:

$$ X(0) = c_3 + c_4 = 0 \implies c_3 = -c_4 ,$$ $\newline$

$$ X(L) = -c_4 \exp{(λ L)} + c_4 \exp{(- λ L)} = 0 \implies c_4 = 0 = c_3. $$ $\newline$

As in case $Λ > 0$ we get $c_3 = 0$ and $c_4 = 0$, this one is also trivial. $\newline$

$\underline{Case}$: $Λ = -λ^2 < 0$

Now we consider equation:

$$ X'' = -λ^2X $$ $\newline$

General solution of such ordinary differential equation is:

$$ X(x) = A \cos{(λ x)} + B \sin{(λ x)}, $$ $\newline$

now we can again use conditions $\eqref{boundary X0}$ and $\eqref{boundary XL}$:

$$ X(0) = A = 0 \implies A = 0, $$ $\newline$

$$ X(L) = A \sin{(λ L)} = 0,  $$ $\newline$

$$ \sin{(λ L)} = 0 \implies λ = \frac{n π}{L}, \;\;\;  n ∈ ℕ. $$ $\newline$

Thus we get the solution for equation $\eqref{equal lambda}$ for function $X(x)$:

$$ X_n(x) = A_n \sin{\left( \frac{n π}{L}x \right)}, \;\;\; n \in ℕ \tag{19}\label{Xn solution}. $$ $\newline$

Now we need to solve equation $\eqref{equal lambda}$ for function $T(t)$. We do it in the same way as in the case of function $X$, thus we get:

$$ T' = -c^2 λ T, $$ $\newline$

$$ T_n(t) = C_n \exp{\left(-\left(\frac{n π c}{L} \right)^2 t \right)}. \tag{20}\label{Tn solution} $$ $\newline$

From $\eqref{separation}$ we know that general solution of diffusion equation can be written as product of functions $X(x)$ and $T(t)$, therefore we have:

$$ y(x,t) = \sum_{n=1}^{\infty} E_n \sin{\left( \frac{n π}{L}x \right)}\exp{\left(-\left(\frac{n π c}{L} \right)^2 t \right)} $$ $\newline$

where $D_n$ is constant such that $E_n = A_nC_n$. This constant can be found using initial condition $\eqref{new initial}$. Thus we have:

$$ y(x,0) = D\exp{\left( - \frac{b}{2c^2}x \right)} f(x) = \sum_{n=1}^{\infty} E_n \sin{\left( \frac{n π}{L}x \right)} \tag{21}\label{find constant}. $$ $\newline$

So the coefficient $E_{n}$ can be written as:

$$ E_{n} = \frac{2}{L} \int_{0}^{L} D\exp{\left( - \frac{b}{2c^2}x \right)} f(x) \sin{\left( \frac{n π}{L}x \right)}  \,dx . \tag{22}\label{En constant} $$ $\newline$

At this point we can finally write down the solution of diffusion-convection equation $\eqref{D-C eq sub}$, as product of function $z(x,t)$ and $y(x,t)$, thus we obtain:

$$ u(x,t) = D \exp{\left( \frac{-b^2}{4c^2}t \right)} \exp{\left(\frac{b}{2c^2}x \right)} \sum_{n=1}^{\infty} E_n \sin{\left( \frac{n π}{L}x \right)}\exp{\left(-\left(\frac{n π c}{L} \right)^2 t \right)} \tag{23}\label{dir solution}, $$ $\newline$

where $E_{n}$ is defined by $\eqref{En constant}$. By proceeding in an analogous way, we can obtain a solution to the problem for the other two boundary conditions, for this reason we will present only solutions.

#### Solution for Neuman Boundary Conditions

Solution of diffusion-convection equation for homogeneous Neumann boundary conditions takes the following form:

$$ u(x,t) = E_0 + \sum_{n=1}^{\infty} E_n \cos{\left( \frac{n π}{L}x \right)}\exp{\left(-\left(\frac{n π c}{L} \right)^2 t \right)} \tag{24}\label{neu solution}, $$ $\newline$

where: $\newline$

$$ E_0 = \frac{1}{L} \int_{0}^{L}f(x) \,dx . \tag{25}\label{E0 neu}, $$ $\newline$

$$ E_n = \frac{2}{L} \int_{0}^{L}f(x) \cos{\left( \frac{n π}{L}x \right)} \,dx  \tag{26}\label{En neu}. $$ $\newline$

#### Solution for Robin Boundary Conditions

Solution of diffusion-convection equation for homogeneous Robin boundary conditions,obtained by proceeding in an analogous way, takes the following form:

$$ u(x,t) = \exp{\left( \frac{-b^2}{4c^2}t \right)} \exp{\left(\frac{b}{2c^2}x \right)} \sum_{n=1}^{\infty} E_n \left( \cos{\left( \frac{a_n}{L}x \right)} + \frac{C_1 - \frac{b}{2c^2}}{a_n}L \sin{\left( \frac{a_n}{L}x \right)} \right) \exp{\left(-\left(\frac{a_n c}{L} \right)^2 t \right)} \tag{27}\label{rob solution}, $$ $\newline$

where: $\newline$

$$ E_n = \frac{\int_{0}^{L} \left( \cos{\left( \frac{a_n}{L}x \right)} + \frac{C_1 - \frac{b}{2c^2}}{a_n}L \sin{\left( \frac{a_n}{L}x \right)} \right)\exp{\left(\frac{b}{2c^2}x \right)}f(x) \,dx}{\int_{0}^{L} \left( \cos{\left( \frac{a_n}{L}x \right)} + \frac{C_1 - \frac{b}{2c^2}}{a_n}L \sin{\left( \frac{a_n}{L}x \right)} \right)^2\,dx}, \tag{28}\label{En rob}$$

and $a_n$ stands for solutions of transcendental equation in the form:

$$ \tan{a} = \frac{L(C_1 - C_2)a}{a^2 + L^2(C_1 - \frac{b}{2c^2})(C_2 + frac{b}{2c^2})}. \tag{29}\label{transcentential}$$

## Finite difference - explicit method

### Discretization of the equation

Having our convection-diffusion equation in such form: 

$$
\frac{\partial u}{\partial t}(x,t) = c^2 \frac{\partial^2u}{\partial x^2}(x,t) - b\frac{\partial u }{\partial x}(x,t) \tag{30}\label{diff conv disc}
$$

we want to approximate the derivatives for time and $x$. We will start from forward difference approximation for time derivative, which is equal to:

$$
\frac{\partial u}{\partial t} = \frac{u_{i}^{n+1} - u_{i}^{n}}{\Delta t}.
$$

The central difference approximation derivative for the first order $x$ can be written as: 

$$
\frac{\partial u }{\partial x} = \frac{u_{i+1}^{n} - u_{i-1}^{n}}{2h},
$$

and last would be symmetric central difference approximation derivative for the second order:

$$
\frac{\partial^2u}{\partial x^2} = \frac{u_{i+1}^{n} - 2u_{i}^{n} + u_{i-1}^{n}}{h^2}.
$$

Now the equation $\eqref{diff conv disc}$, after plugging in obtained result becomes:

$$
\frac{u_{i}^{n+1} - u_i^n}{\Delta t} = c^2 \frac{u_{i+1}^{n} - 2u_{i}^{n} + u_{i-1}^{n}}{h^2} - b \frac{u_{i+1}^{n} - u_{i-2}^{n}}{2h}.
$$

We want to derive the formula for $u_i^{n+1}$. Thus:

$$
u_{i}^{n+1} = \left(\frac{c^2\Delta t}{h^2} - \frac{b \Delta t}{2h} \right)u_{i+1}^{n} + \left(1 - \frac{2c^2 \Delta t} {h^2}\right)u_{i}^{n} + \left(\frac{c^2\Delta t}{h^2} + \frac{b \Delta t}{2h} \right)u_{i-1}^{n}.
$$

Let us define: 

$$
\alpha = \frac{c^2 \Delta t}{h^2} \\
\beta = \frac{b \Delta t}{2h}.
$$

Hence, we will have: 

$$
u_{i}^{n+1} =  (\alpha - \beta) u_{i+1}^{n} + (1-2\alpha)u_{i}^{n} + (\alpha + \beta)u_{i-1}^{n}
$$

### Stability
In case of the stability, we know that this method would be numerically stable if and only if $\frac{\Delta t}{h^2} < \frac1{2}$.

Implementation of above method contains matrix A, which is described in details in Black-Scholes section.

```python
def A_matrix(b, c, dt, dx, N):
    alpha = c*dt/(dx**2)
    beta = b*dt/(2*dx)

    ds = np.diag(1 - 2*alpha*np.ones(N-1))

    ns = np.arange(1, N)
    us = np.diag((alpha+beta)*np.ones(N-1))
    us = np.roll(us, 1, axis=1)
    us[N-2, 0] = 0

    ns = np.arange(1, N)
    ls = np.diag((alpha-beta)*np.ones(N-1))
    ls = np.roll(ls, -1, axis=1)
    ls[0, N-2] = 0

    return ds + us + ls

def convection_diffusion_finite_difference(b, c, t_max, x_max, N, M):
    dt = t_max / M

    # assert dt > 0 and dt < 1 / (sigma**2 * (N-1) + 0.5 * r), "Stability condition is not satisfied"

    dx = x_max / N

    alpha = c*dt/(dx**2)
    beta = b*dt/(2*dx)
    A = A_matrix(b, c, dt, dx, N)

    l_0 = (alpha + beta)
    u_N = (alpha - beta)

    grid = np.zeros(shape=(M+1, N+1))

    x = np.linspace(0, x_max, N+1)
    grid[0] = np.sin(np.pi*x)
    grid[:, 0] = 0
    grid[:, N] = 0

    for m1 in range(1, M+1):
        z_m = np.zeros(N-1)
        z_m[0] = l_0 * grid[m1-1][0]
        z_m[-1] = u_N * grid[m1-1][N]
        grid[m1][1:-1] = grid[m1-1][1:-1] @ A + z_m

    return grid
```

## Black-Scholes equation



The Black-Scholes equation is a partial differential equation, which has begun a new era of modern financial mathematics in 1973. Generally speaking it expanded the knowledge of contemporary investments. The derived formula for this equation, which we are going to look into, allows us to price various options. The most popular, and the ones we will consider are the European call option and European put option. In the first part we are going to present an analytic solution for the first mentioned option. [[6](#bib)]

Let us begin with the basic form of our equation. The Black-Scholes partial differential equation can we written in the following way:

$$\frac{\partial V}{\partial t} + \frac1{2}\sigma^2 S^2\frac{\partial^2 V}{\partial S^2} + rS\frac{\partial V}{\partial S} - rV = 0, \tag{31}\label{black scholes} $$ $\newline$
where: $\newline$
$σ$ - volatility, $\newline$
$r$ - risk-free interest rate, $\newline$
$S$ - stock price, $\newline$
$V$ - value of the considered option.


The formula stated in $\eqref{black scholes}$ is written without choosing the option we want to analyse. In our case we want to find a solution for the European call option, so we need to rewrite the equation with particular $V$. From now it would become the value of European call option $C(S,t)$. Hence, we can write our equation as:

$$ \frac{∂C}{∂t} + \frac{1}{2} \sigma^2 S^2 \frac{∂^2C}{∂S^2} + rS\frac{∂C}{∂S} - rC = 0. \tag{32}\label{partial derivative call}
$$

#### Boundary conditions

From the knowledge of pricing derivatives, we actually know some basic assumptions, which will help us in covering the boundary conditions. Otherwise the equation will not have a unique solution. One of them is simply the point that the call option is worthless for $S = 0$ and the second one for $S \rightarrow \infty$, the value of the option becomes the asset. Based on the European call option the following statements occur: [[5](#bib)]$\newline$

$$
C(0,t) = 0, \hspace{0.4em} \text{for all} \hspace{0.4em} t \geq 0, \\ 
C(S,t) \sim S \hspace{0.4em} \text{as} \hspace{0.4em} S \to \infty \hspace{0.4em} \text{for all} \hspace{0.4em} t > 0, \tag{33}\label{initials} \\
$$
where: $\newline$
- $E$ - strike price, $\newline$
- $T$ - expiry date. $\newline$

#### Final condition

And a very important fact is that our final condition is equal to (payoff function for European call option):

$$
C(S,T) = max(S-E, 0), \hspace{0.4em} \text{for all} \hspace{0.4em} S \geq 0. 
$$

### Analytical solution

By having our equation in form given in $\eqref{partial derivative call}$, we would like to solve it in the analytical way. Just by looking at it we can see that it is clearly in a backward form and also that it somehow resembles the diffusion equation with more terms. Hence the first step to find the solution is to basically transform the Black-Scholes partial differential equation for European call option into the one-dimensional diffusion equation.

To do so we need to apply the change of variables transformation. This will make the variables dimensionless and we will obtain a forward parabolic equation. Hence, we have to introduce: [[3](#bib)] $\newline$

$$\tau = \frac{2(T - t)}{\sigma^2}, \\
S = Ee^{y}, \tag{34} \label{new variables}\\
C(S,t) = Ev(y,\tau). $$ $\newline$

Partial derivatives in equation $\eqref{partial derivative call}$ are going to be replaced by the newly defined variables in equation $\eqref{new variables}$. Hence, we need to calculate the derivatives. When it comes to $\frac{∂C}{∂S}$, we now obtain the following result:  

$$
\frac{∂v}{∂y} = \frac1{E} \frac{\partial C}{\partial S} \frac{∂S}{∂y} = \frac1{E} \frac{\partial C}{\partial S} E e^{y} \Rightarrow \frac{∂C}{∂S} = e^{-y} \frac{∂v}{∂y}.
$$

Moving on to the other two partial derivatives, we now have:

$$
\frac{∂v}{∂\tau} = \frac1{E} \frac{∂C}{∂t} \frac{∂t}{∂\tau} = \frac1{E} \frac{∂C}{∂t} \left( \frac{-2}{\sigma^2} \right) \Rightarrow \frac{\partial C}{\partial t} = - \frac{\sigma^2 E}{2} \frac{∂v}{∂\tau}.
$$
And $\newline$
$$
\frac{∂^2v}{∂y^2} = e^{y} \frac{∂C}{∂S} + e^{y} \frac{∂^2C}{∂S^2} \frac{\partial S}{\partial y} = e^{y} \frac{∂C}{∂S} + e^{y} \frac{∂^2C}{∂S^2} Ee^{y} = \\
= e^{y} e^{-y} \frac{∂v}{∂y} + e^{y} \frac{∂^2C}{∂S^2} Ee^{y} \Rightarrow \frac{∂^2C}{∂S^2} = \frac{e^{-2y}}{E} \left( \frac{∂^2v}{∂y^2} - \frac{∂v}{∂y} \right).
$$

Now we need to come back to the previous equation in $\eqref{partial derivative call}$ and plug in our results for the introduced variables. Hence, we have the form of our analysed equation given as:
$$
- \frac{\sigma^2 E}{2} \frac{∂v}{∂\tau} + \frac{\sigma^2 E^2}{2} e^{2y} \frac{e^{-2y}}{E} \left(\frac{∂^2v}{∂y^2} - \frac{∂v}{∂y}\right) + rE e^{y} e^{-y} \frac{∂v}{∂y} - rEv = 0. \tag{35}\label{transforming bs}
$$

Now after appropriate transformations, we obtain the result $\eqref{transforming bs}$ in the form:

$$
- \frac{\sigma^2 E}{2} \frac{∂v}{∂\tau} + \frac{\sigma^2 E}{2} \left(\frac{∂^2v}{∂y^2} - \frac{∂v}{∂y}\right) + rE \frac{∂v}{∂y} - rEv = 0.
$$

For $E > 0$ we just get the equation in even simpler form. After getting rid of the strike price $E$ we just need to make some basic transformations. Hence:

$$
\begin{equation}
- \frac{\sigma^2}{2} \frac{∂v}{∂\tau} + \frac{\sigma^2}{2} \frac{∂^2v}{∂y^2} - \frac{\sigma^2}{2} \frac{∂v}{∂y} + r\frac{∂v}{∂y} - rv = \\
= \frac{\sigma^2}{2} \frac{∂^2v}{∂y^2} + \left(r-\frac{\sigma^2}{2}\right)\frac{∂v}{∂y} - \frac{\sigma^2}{2} \frac{∂v}{∂\tau} -rv = 0. \tag{36}\label{some evaluations}
\end{equation}
$$

Next step in our solution is to define the new constant, to get the more convenient form of our equation. For this purpose let:

$$
k = \frac{2r}{\sigma^2}. \tag{37}\label{ka}
$$

And now we multiply our equation $\eqref{some evaluations}$ by $\frac{2}{\sigma^2} > 0$. We obtain:

$$
\frac{∂^2v}{∂y^2} + (k-1)\frac{∂v}{∂y} - \frac{\partial v}{\partial \tau} - kv = 0 \Rightarrow \frac{\partial v}{\partial \tau} = \frac{∂^2v}{∂y^2} + (k-1)\frac{∂v}{∂y} - kv. \tag{38}\label{more evals}
$$

As we can see in the equation $\eqref{more evals}$ we get the result, which is the forward parabolic equation. Our final condition for $C(S,T)$ becomes the initial condition for $v$. Hence, we have: [[3](#bib)]
$$
v(y,0) = max\left(e^{y}-1, 0\right).
$$

When it comes to boundary conditions given in $\eqref{initials}$, we now have: [[3](#bib)]

$$\left\{\begin{array}{ll} 
v(y,\tau) = 0, \hspace{0.4em} y \rightarrow -\infty, \\
v(y,\tau) \sim e^{y} - e^{-r\tau} \sim e^{y}, \hspace{0.4em} y \rightarrow \infty. 
  \end{array} \right.
$$

If we look at the formula in equation $\eqref{more evals}$ we can see that it is getting closer to our destination, which is to get the form of the diffusion equation. Our next step is going to be the reduction of the right hand side. For this purpose we need to define:

$$
v(y, \tau) = e^{\alpha y + \beta \tau} w(y,\tau), \tag{39}\label{v y tau} \\
$$
where: $\newline$
- $\alpha, \beta$ - constants.

Now with this substitution the equation $\eqref{more evals}$ becomes:

$$
e^{\alpha y + \beta \tau} \left(\beta w + \frac{\partial w}{\partial \tau} \right) = e^{\alpha y + \beta \tau} \left[\left(\alpha^2w + 2\alpha \frac{\partial w}{\partial y} + \frac{\partial^2 w}{\partial y^2}\right) + (k-1) \left(\alpha w + \frac{\partial w}{\partial y} \right) - kw \right]. \tag{40}\label{alfa beta}
$$

After division and appropriate transformations of the equation $\eqref{alfa beta}$, we obtain:

$$
\beta w + \frac{\partial w}{\partial \tau} = \alpha^2w + 2\alpha \frac{\partial w}{\partial y} + \frac{\partial^2 w}{\partial y^2} + (k-1) \left(\alpha w + \frac{\partial w}{\partial y} \right) - kw \\
\frac{\partial w}{\partial \tau} = \frac{\partial^2 w}{\partial y^2} + \left(2\alpha + k - 1 \right) \frac{\partial w}{\partial y} + \left(\alpha^2 + \alpha (k-1) - \beta - k \right)w.
$$

We want to get rid of the partial derivative $\frac{\partial w}{\partial y}$. Hence we need to define $\alpha$ for which the part ($2\alpha + k - 1$) would be equal to $0
$. Then:

$$
2\alpha + k - 1 = 0 \Rightarrow \alpha = -\frac{1}{2}(k-1). \tag{41}\label{alfa}
$$
Now, by knowing $\alpha$ we can find the formula for $\beta$ and cancel out the part with $w$. Hence, after basic transformations, we have:

$$
\beta = \alpha^2 + \alpha(k-1) - k = \frac1{4} (k-1)^2 - \frac1{2}(k-1)^2 - k = - \frac1{4}(k-1)^2 - k = \\
= - \frac1{4}(k^2 - 2k + 1) - k = - \frac1{4}k^2 + \frac1{2}k - \frac1{4} - k = - \frac1{4}(k+1)^2. \tag{42}\label{beta}
$$

For $\alpha$ and $\beta$ given respectively in equations $\eqref{alfa}$ and $\eqref{beta}$, we finally reached the formula:

$$
\frac{\partial w}{\partial \tau} = \frac{\partial^2 w }{\partial y^2}, \tag{43}\label{one dimensional diff eq}
$$

which is a one-dimensional diffusion equation with the initial condition given as:

$$
w(y,0) = w_0(y) = e^{\frac1{2}(k-1)y}v(y,0) = max\left(e^{\frac1{2}(k+1)y}- e^{\frac1{2}(k-1)y}, 0 \right). \tag{44}\label{wy0}
$$

The solution to equation $\eqref{one dimensional diff eq}$ is a well known formula for $\tau > 0$ and $-\infty  < y < \infty$. The formula is equal to: [[7](#bib)]

$$
w(y, \tau) = \frac1{2\sqrt{\pi \tau}} \int\limits_{-\infty}^{\infty} w_{0}(s) e^{\frac{-(y-s)^2}{4\tau}} ds, \tag{45}\label{wytau} \\
$$

where $u_{0}$ is of the form given in equation $\eqref{wy0}$.


What is left is just to evaluate the integral in $\eqref{wytau}$. For convenience we will make a change of variable, so let us introduce:

$$
z = \frac{(s-y)}{\sqrt{2\tau}}.
$$
And after determining $s$ and $y$ we can plug it into our formula in $\eqref{wytau}$. Hence, we obtain:

$$
w(y, \tau) = \frac1{\sqrt{2\pi}} \int\limits_{-\infty}^{\infty} w_{0}(z\sqrt{2\tau} + y) e^{-\frac1{2}z^2}dz.
$$

We know that our initial condition is derived in $\eqref{wy0}$, so next step would be to write it for the newly defined $z$ and plug it into the integral. Then:

$$
w(y, \tau) = \frac1{\sqrt{2\pi}} \int\limits_{-\infty}^{\infty} max\left(e^{\frac1{2}(k+1)(z\sqrt{2\tau} + y)} - e^{\frac1{2}(k-1)(z\sqrt{2\tau} + y)}, 0 \right) e^{-\frac1{2}z^2}dz.
$$
Here to evalutate that integral we need to change the lower limit integration. The formula will be given as:
$$
w(y, \tau) = \frac1{\sqrt{2\pi}} \int\limits_{-\frac{y}{\sqrt{2\tau}}}^{\infty} e^{\frac1{2}(k+1)(z\sqrt{2\tau} + y)} - e^{\frac1{2}(k-1)(z\sqrt{2\tau} + y)} e^{-\frac1{2}z^2}dz.  
$$
In order to get the final result it would be easier to split the formula to two integrals. After that we will calculate them separately. Hence:
$$
w(y, \tau) = \frac1{\sqrt{2\pi}} \left(\int\limits_{-\frac{y}{\sqrt{2\tau}}}^{\infty} e^{\frac1{2}(k+1)(z\sqrt{2\tau} + y)} e^{-\frac1{2}z^2}dz - \int\limits_{-\frac{y}{\sqrt{2\tau}}}^{\infty} e^{\frac1{2}(k-1)(z\sqrt{2\tau} + y)} e^{-\frac1{2}z^2}dz \right). \tag{46}\label{to calc I1 and I2}
$$

Let us concentrate on the left hand side. We are going to perform some calculations to get the final formula. 

$$
I_{1} = \frac1{\sqrt{2\pi}} \int\limits_{-\frac{y}{\sqrt{2\tau}}}^{\infty} e^{\frac1{2}(k+1)(z\sqrt{2\tau} + y)} e^{-\frac1{2}z^2}dz = \frac{e^{\frac1{2}(k+1)y}}{\sqrt{2\pi}} \int\limits_{-\frac{y}{\sqrt{2\tau}}}^{\infty} e^{\frac1{2}(k+1)(z\sqrt{2\tau})-\frac1{2}z^2}dz
$$

As we could see previously that the impression under the integral was quite similar to the cumulative function for the normal distribution. In that form after merging exponential functions into one formula, we want to obtain the expression in the form $-\frac1{2}(...)^2$ and substract what is left, being the constant, that we can pull out in front of our integral in the next step. To do so we need to modify the exponential function in the following way:

$$
I_{1} = \frac{e^{\frac1{2}(k+1)y}}{\sqrt{2\pi}} \int\limits_{-\frac{y}{\sqrt{2\tau}}}^{\infty} e^{-\frac1{2}\left(z - \frac1{2}(k+1)\sqrt{2\tau}\right)^2 + \frac1{4}(k+1)^2 \tau}dz = \\
= \frac{e^{\frac1{2}(k+1)y + \frac1{4}(k+1)^2 \tau}}{\sqrt{2\pi}} \int\limits_{-\frac{y}{\sqrt{2\tau}}}^{\infty} e^{-\frac1{2}\left(z - \frac1{2}(k+1)\sqrt{2\tau}\right)^2}dz.
$$

What is left is just to substitute $p = z-\frac1{2}(k+1)\sqrt{2\tau}$ and change the integration limits. After that we get the final formula:

$$
I_{1} = \frac{e^{\frac1{2}(k+1)y + \frac1{4}(k+1)^2 \tau}}{\sqrt{2\pi}} \int_{-\frac{y}{\sqrt{2\tau}}-\frac1{2}(k+1)\sqrt{2\tau}}^{\infty} e^{-\frac1{2}p^2}dp = e^{\frac1{2}(k+1)y + \frac1{4}(k+1)^2 \tau} \Phi(d_{1}), \\
$$
where
$$
d_{1} = \frac{y}{\sqrt{2\tau}}+\frac1{2}(k+1)\sqrt{2\tau},
$$
and 
$$
\Phi(d_{1}) = \frac1{\sqrt{2\pi}} \int\limits_{-\infty}^{d_1} e^{-\frac1{2}p^2}dp.
$$ 

Similar calculations are needed to obtain the result for the second integral. The only difference is that we have $(k-1)$ expect of $(k+1)$.  Hence we are going to skip them and proceed to the final formula, which is given by:

$$
I_{2} = e^{\frac1{2}(k-1)y + \frac1{4}(k-1)^2 \tau} \Phi(d_{2}), \\
$$

where

$$
d_{2} = \frac{y}{\sqrt{2\tau}} + \frac1{2}(k-1)\sqrt{2\tau}.
$$

Moving back to equation $\eqref{to calc I1 and I2}$, we can derive the final formula for $w(y, \tau)$:

$$
w(y, \tau) = I_{1} - I_{2} = e^{\frac1{2}(k+1)y + \frac1{4}(k+1)^2 \tau} \Phi(d_{1}) - e^{\frac1{2}(k-1)y + \frac1{4}(k-1)^2 \tau} \Phi(d_{2}).
$$

Now we need to go back to the formula we obtained in $\eqref{more evals}$ and after that we will plug everything we derived during whole process including $\alpha$ and $\beta$ into the starting formula in $\eqref{new variables}$.

$$
v(y, \tau) = e^{-\frac1{2}(k-1)y - \frac{1}{4}(k+1)^2\tau}w(y, \tau) = e^{-\frac1{2}(k-1)y - \frac{1}{4}(k+1)^2\tau} \left(I_{1} - I_{2} \right).
$$

And now going back to $\eqref{new variables}$, we obtain that:

$$
Ev(y,\tau) = \\
= E e^{-\frac1{2}(k-1)y - \frac{1}{4}(k+1)^2\tau} \left(e^{\frac1{2}(k+1)y + \frac1{4}(k+1)^2 \tau} \Phi(d_{1}) - e^{\frac1{2}(k-1)y + \frac1{4}(k-1)^2 \tau} \Phi(d_{2})\right) = \\
=  E e^{-\frac1{2}(k-1)y - \frac{1}{4}(k+1)^2\tau + \frac1{2}(k+1)y + \frac1{4}(k+1)^2 \tau} \Phi(d_{1}) -  E e^{-\frac1{2}(k-1)y - \frac{1}{4}(k+1)^2\tau + \frac1{2}(k-1)y + \frac1{4}(k-1)^2 \tau} \Phi(d_{2}) = Ee^{y}\Phi(d_{1}) - Ee^{-k \tau}.
$$

We know the formula for $\tau$ and to get $y$ we need to determine it from equation $\eqref{new variables}$. Based on that we have:

$$
S = Ee^{y} \Rightarrow y = \ln\left({\frac{S}{E}}\right). \tag{47}\label{derived y}
$$

Hence our final result for price of the European call option after plugging in all defined variables is equal to:

$$
C(S,t) = E \frac{S}{E} \Phi(d_{1}) - Ee^{-\frac{2r}{\sigma^2} \frac{\sigma^2(T-t)}{2}} = S \Phi(d_{1}) - Ee^{r(T-t)}\Phi(d_{2}), \tag{48}
$$

where the formula for $d_{1}$ after plugging in $k$ from equation $\eqref{ka}$ ,$\tau$  defined in equation $\eqref{new variables}$ and $y$ from equation $\eqref{derived y}$ is equal:

$$
d_{1} = \frac{y}{\sqrt{2\tau}}+\frac1{2}(k+1)\sqrt{2\tau} = \frac{\ln\left({\frac{S}{E}}\right)}{\sqrt{2\frac{\sigma^2(T-t)}{2}}} + \frac1{2}\left(\frac{2r}{\sigma^2} + 1\right) \sqrt{2\frac{\sigma^2(T-t)}{2}} = \\
= \frac{\ln\left({\frac{S}{E}}\right) + (r + \frac{\sigma^2}{2}) (T-t)}{\sigma \sqrt{(T-t)}}.
$$

And after the same transformations for $d_{2}$, we obtain:

$$
d_{2} = \frac{y}{\sqrt{2\tau}} + \frac1{2}(k-1)\sqrt{2\tau} = \frac{\ln\left({\frac{S}{E}}\right)} {\sqrt{2\frac{\sigma^2(T-t)}{2}}} + \frac1{2}\left(\frac{2r}{\sigma^2} - 1\right) \sqrt{2\frac{\sigma^2(T-t)}{2}} = \frac{\ln\left({\frac{S}{E}}\right) + (r - \frac{\sigma^2}{2}) (T-t)}{\sigma \sqrt{(T-t)}} = d_{1} - \sigma \sqrt{T}.
$$

### European put option
We have the result for the European call option, but in our numerical solution we would like to show also the put option. To get that we should follow the same calculations with the difference in the final condition and boundary conditions. Thus for the European put option we consider the final condition as:

$$
P(S,t) = max(E-S, 0) \hspace{0.4em} \text{for all} \hspace{0.4em} S \geq 0.
$$

And the boundary conditions:

$$
P(0,t) = Ee^{-r\tau}, \hspace{0.4em} \text{for all} \hspace{0.4em} t \geq 0, \\ 
P(S,t) \sim 0 \hspace{0.4em} \text{as} \hspace{0.4em} S \to \infty \hspace{0.4em} \text{for all} \hspace{0.4em} t > 0
$$

The result we would obtain for the price of the European put option is equal to:

$$
P(S,t) = Ee^{-r(T-t)} \Phi (-d_{2}) - S\Phi(-d_{1}). 
$$

### Implementation of analytical solution

```python
# analytical method
def bs_price(E, r, sigma, T, S, is_call=True):
    d1 = (np.log(S / E) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    if is_call:
        price = S * stats.norm.cdf(d1) - E * np.exp(-r * T) * stats.norm.cdf(d2)
    else:
        price = E * np.exp(-r * T) * stats.norm.cdf(-d2) - S * stats.norm.cdf(-d1)

    return price
```

### Finite difference - discretization of equation
[[4](#bib)]

---
#### Change of variables
To transform the terminal condition of the option price into an initial condition, we utilize a time inverse Black-Scholes equation. This involves introducing a change of variable, where we define a new time variable $\tau$ such that $t = T - \tau$. By applying this transformation, we can rewrite the original Black-Scholes equation as follows:

$$\frac{\partial V}{\partial t} - \frac{1}{2} \sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} - r S \frac{\partial V}{\partial t} + rV = 0. \tag{49}\label{49}$$

#### Space discretization
The price of the underlying asset is considered in the unbounded interval $[0, \infty)$, but we introduce an artificial limit, $S_{max}$, which is typically chosen to be around three to four times the exercise price $E$.

The price range $[0, S_{max}]$ is divided into N equally sized subintervals of length $\Delta S$. This allows us to approximate the space $[0, S_{max}] \times [0, T]$ using a grid:

$$(n\Delta S, m\Delta t) \in [0, N \Delta S] \times [0, M \Delta T],$$

where n = 0, $\dots$, N and m = 0, $\dots$, M. The grid represents the stock price at different time points.

#### Initial conditions

For call option the final condition due to the change of variables it becomes initial condition:

$$C(S, 0) = max(S-E, 0)$$


Similarly for the put option due to the change of vairables it is changed to the initial condition:

$$P(S, 0) = max(E-S, 0)$$


##### The explicit method 

For approximation of derivatives we will use following differences:

Forward difference approximation for the time derivative:

$$\frac{\partial V}{\partial t} = \frac{v_n^{m+1} - v_n^m}{\Delta t} + O(\Delta t), \tag{50}$$

Central difference approximation for the first order S derivative:

$$\frac{\partial V}{\partial S} = \frac{v_{n+1}^{m} - v_{n-1}^m}{2\Delta S} + O((\Delta S)^2), \tag{51}$$

Symetric central difference approximation for the
second order S derivative:

$$\frac{\partial^2 V}{\partial S^2} = \frac{v_{n+1}^{m} - 2v_{n}^{m} + v_{n-1}^m}{(\Delta S)^2} + O((\Delta S)^2), \tag{52}$$

From equation $\eqref{49}$ we have:
$$\frac{v_n^{m+1} - v_n^m}{\Delta t} - \frac{1}{2} \sigma^2 S^2  \frac{v_{n+1}^{m} - 2v_{n}^m + v_{n-1}^m}{(\Delta S)^2} - r S \frac{v_{n+1}^{m} - v_{n-1}^m}{2\Delta S} + r v_n^m \tag{53}\label{53}$$

Note that $S = n \Delta S$ and from equation $\eqref{53}$ we can determine $v_n^{m+1}$:

$$v_n^{m+1} = \frac{1}{2}(\sigma^2 n^2 \Delta t - rn \Delta t)v_{n-1}^m + (1 - \sigma^2 n^2 \Delta t - r \Delta t) v_n^m + \frac{1}{2} (\sigma^2 n^2 \Delta t + r n \Delta t ) v_{n+1}^m \tag{54}$$

where $n = 1, \dots, N-1, m= 0, \dots, M-1$ 


It can be seen that the term $v_n^{m+1}$ for time step $m+1$ is calculated only using the terms $v_{n-1}^{m}, v_n^{m}, v_{n+1}^m$ from the previous time step $m$. 


Let's denote:
$$\alpha = \sigma^2 \Delta t  \tag{55}\label{alpha betaa imp ex},$$
$$\beta = r \Delta t.$$

We obtain:

$$v_n^{m+1} = \frac{1}{2} (\alpha n^2 - \beta n) v_{n-1}^m + (1-\alpha n^2 - \beta) v_{n}^m + \frac{1}{2} (\alpha n^2 + \beta n)v_{n+1}^m,$$

for $n=1,\dots,N-1, m=0,\dots, M-1$.


The matrix is introduced:



$$A = \begin{bmatrix}
d_1 & u_2 & \cdots & 0 & 0 & 0\\
l_1 & d_2 & u_3 &  & \vdots & \vdots\\
0 & l_2 & \ddots & \vdots & &  \\
0 & 0 &  &  & \ddots & 0 \\
0 & 0 &  & \ddots & \ddots & u_{N-1} \\
0 & 0 &  & 0 & l_{N-2} & d_{N-1} \\
\end{bmatrix} \tag{56}\label{matrix A}$$


where $A \in \mathbb{R}^{(N-1) \times (N-1)}$,

To find numericaly optimal formula we introduce vectors of size $N-1$:

$$v^{m} = \begin{pmatrix} 
v_{1}^{m} \\
v_{2}^{m} \\
\vdots \\
v_{N-1}^{m}
\end{pmatrix}, z^m = \begin{pmatrix} 
l_0 v_0^m \\
0 \\
\vdots \\
0 \\
u_N v^m_N
\end{pmatrix}$$

and

$$
\begin{align}
& d_n = 1 - \alpha n^2 - \beta  & n = 1, \dots, N-1  \\
& u_n = \frac{1}{2}\left( \alpha (n-1)^2 + \beta (n-1) \right) & n = 2, \dots, N, \phantom{-2 }   \\
& l_n = \frac{1}{2}\left( \alpha (n+1)^2 - \beta (n+1) \right) & n = 0, \dots, N-2 
\end{align}
$$

Now we can obtain formula for solution approximation at each time step $m+1$:

$$v^{m+1} = A v^m + z^m.$$

#### Stability

This explicit method will be numerically stable whenever $\frac{\Delta t}{(\Delta S)^2} < \frac1{2}$.


#### Fully implicit method

After using the explicit method we would now concenrate on the fully implicit method, which differs from the previous one by the fact that here we will use backward difference approximation for the time derivative. Hence, we now have:

$$
\frac{\partial V}{\partial t} = \frac{v_{n}^{m+1} - v_{n}^{m}}{\Delta t} + O(\Delta t).
$$

The derivative for $S$ is equal to:

$$
\frac{\partial V}{\partial S} = \frac{v_{n+1}^{m+1} - v_{n-1}^{m+1}}{2\Delta S} + O((\Delta S^2)),
$$

and for the second order derivative, we obtain:

$$
\frac{\partial^2 V}{\partial S^2} =  \frac{v_{n+1}^{m+1} - 2v_{n}^{m+1} + v_{n-1}^{m+1}}{(\Delta S)^2} + O((\Delta S)^2).
$$

The equation $\eqref{49}$ after plugging in all derivatives becomes:

$$
\frac{v_{n}^{m+1} - v_{n}^{m}}{\Delta t} - \frac1{2}\sigma^2 S^2 \frac{v_{n+1}^{m+1} - 2v_{n}^{m+1} + v_{n-1}^{m+1}}{(\Delta S^2)} - rS \frac{v_{n+1}^{m+1} - v_{n-1}^{m+1}}{2\Delta S} + rv_{n}^{m+1 }. \tag{57}\label{implicit}
$$

From equation $\eqref{implicit}$ we can determine term $v_{n}^{m}$, which from time step $m$ is evaluated using the terms $v_{n-1}^{m+1}$, $v_{n}^{m+1}$, $v_{n+1}^{m+1}$ from time step $m+1$. Hence:

$$
v_{n}^{m} = \frac1{2}(rn\Delta t - \sigma^2 n^2 \Delta t)v_{n-1}^{m+1} + \left(1 + r\Delta t + \sigma^2n^2\Delta t\right)v_{n}^{m+1} - \frac1{2}\left(rn\Delta t + \sigma^2n^2\Delta t \right)v_{n+1}^{m+1}, \tag{58}\label{vnm}
$$

where $n = 1, \dots, N-1, m= 0, \dots, M-1$.

Using provided notations for $\alpha$ and $\beta$ in $\eqref{alpha betaa imp ex}$ the equation $\eqref{vnm}$ is now equal to:

$$
v_{n}^{m} = \frac1{2} \left(\beta n - \alpha n^2 \right)v_{n-1}^{m+1} + \left(1 + \beta + \alpha n^2 \right)v_{n}^{m+1} - \frac1{2}\left(\beta n + \alpha n^2 \right)v_{n+1}^{m+1},
$$

where $n = 1, \dots, N-1, m = 0, \dots, M-1$.

We can now use the matrix $A$, that we introduced in equation $\eqref{matrix A}$. For this matrix we have the following formula for $v^{m+1}$:

$$
v^{m+1} = \begin{pmatrix} 
v_{1}^{m+1} \\
v_{2}^{m+1} \\
\vdots \\
v_{N-1}^{m+1}
\end{pmatrix}, b^m = \begin{pmatrix}
v_1^{m} \\
v_2^{m} \\
\vdots \\
v_{N-2}^{m} \\
v_{N-1}^m 
\end{pmatrix}
- \begin{pmatrix} 
l_0 v_0^{m+1} \\
0 \\
\vdots \\
0 \\
u_N v^{m+1}_N
\end{pmatrix},$$

where $v_{m+1}$, $v^{m}$, $b^{m}$ $\in \mathbf{R}^{N-1}$ and 

$$
\begin{align}
& d_n = 1 + \beta + \alpha n^2, & n = 1, \dots, N-1  \\
& u_n = -\frac{1}{2}\left( \beta (n-1) + \alpha (n-1)^2 \right), & n = 2, \dots, N, \phantom{-2}   \\
& l_n = \frac{1}{2}\left( \beta (n+1) - \alpha (n+1)^2 \right), & n = 0, \dots, N-2.
\end{align}
$$

At each time step $m + 1$ the approximate solution can be obtained by solving the system of equations:

$$
Av^{m+1} = b^m.
$$


We construct a matrix $A$ of size $(N-1) \times (N-1)$ that represents the coefficients in the system of equations. The diagonal elements of A correspond to the central differences in the spatial direction, while the off-diagonal elements account for the differences in the neighboring spatial points.

We set the lower boundary condition at each time step by considering the payoff of the call option at each stock price. The upper boundary condition represents the expected value of the call option at expiration, which is determined by the Black-Scholes formula.

To solve the system of equations, we use matrix inversion. At each time step, we calculate the right-hand side of the equation as a vector b, which depends on the values at the previous time step. We then solve the system $A \cdot x = b$ for the vector $x$, which represents the option prices at the current time step.

By iterating over time steps, we calculate the option prices for the entire grid. The resulting grid contains the option prices at different stock prices and time points, providing an approximation of the call option prices based on the Black-Scholes equation.


### Explicit method

```python
def A_matrix(r, sigma, dt, N):
    alpha = sigma**2 * dt
    beta = r * dt

    ds = np.diag(1 - (np.arange(1, N)**2) * alpha - beta)

    ns = np.arange(1, N)
    us = np.diag(0.5 * (alpha * (ns**2) + beta * (ns)))
    us = np.roll(us, 1, axis=1)
    us[N-2, 0] = 0

    ns = np.arange(1, N)
    ls = np.diag(0.5 * (alpha * ((ns)**2) - beta * (ns)))
    ls = np.roll(ls, -1, axis=1)
    ls[0, N-2] = 0

    return ds + us + ls

def bs_explicit(E, r, sigma, T, S_MAX, N, M, is_call=True):
    dt = T / M

    assert dt > 0 and dt < 1 / (sigma**2 * (N-1) + 0.5 * r), "Stability condition is not satisfied"

    ds = S_MAX / N

    A = A_matrix(r, sigma, dt, N)

    s = np.linspace(0, S_MAX, N+1)

    alpha = sigma**2 * dt
    beta = r * dt

    l_0 = 0.5 * (alpha - beta)
    u_N = 0.5 * (alpha * ((N-1)**2) + beta * (N-1))
    grid = np.zeros(shape=(M+1, N+1))

    # initial conditions
    S = np.linspace(0, S_MAX, N+1)
    if is_call:
        grid[:, N] = S_MAX
        grid[0] = np.maximum(S-E, 0)
    else:
        grid[:, 0] = E*np.exp(-r*ts)
        grid[:, N] = 0
        grid[0] = np.maximum(E-S, 0)


    for m1 in range(1, M+1):
        z_m = np.zeros(N-1)
        z_m[0] = l_0 * grid[m1-1][0]
        z_m[-1] = u_N * grid[m1-1][N]
        grid[m1][1:-1] = A @ grid[m1-1][1:-1] + z_m

    return grid
```

### Plotting the solution

```python
r = 0.1
sigma = 0.4
T = 0.25
E = 10
S_MAX = E*4
M = 40000
N = 300
call_finite_diff = bs_explicit(E, r, sigma, T, S_MAX, N, M)


S = np.linspace(0, S_MAX, N+1)

call_theoretical = bs_price(E, r, sigma, T, S[1:])


plt.plot(S, call_finite_diff[-1], label="Finite difference")
plt.plot(S[1:], call_theoretical, label="Analytical form")
plt.axvline(x = E, color = 'b', linestyle='--' , label = 'Strike price')
plt.legend()
```

### Finite difference - the implicit method

```python
def A_matrix_implicit(r, sigma, dt, N):
    alpha = sigma**2 * dt
    beta = r * dt

    ds = np.diag(1 + beta + alpha * (np.arange(1, N)**2))

    ns = np.arange(2, N+1)
    us = np.diag(-0.5 *(beta * (ns-1) + alpha * ((ns-1)**2)))
    us = np.roll(us, 1, axis=1)
    us[N-2, 0] = 0

    ns = np.arange(0, N-1)
    ls = np.diag(0.5 * (beta * (ns-1) - alpha * (ns+1)**2))
    ls = np.roll(ls, -1, axis=1)
    ls[0, N-2] = 0
    return ds + us + ls

def bs_implicit(E, r, sigma, T, S_MAX, N, M, is_call=True):
    dt = T / M

    assert dt > 0 and dt < 1 / (sigma**2 * (N-1) + 0.5 * r), "Stability condition is not satisfied"

    ds = S_MAX / N

    A = A_matrix_implicit(r, sigma, dt, N)

    alpha = sigma**2 * dt
    beta = r * dt

    l_0 = 0.5 * (beta - alpha)
    u_N = -0.5 * (alpha * ((N-1)**2) + beta * (N-1))

    grid = np.zeros(shape=(M+1, N+1))
    S = np.linspace(0, S_MAX, N+1)
    ts = np.linspace(0, T, M+1)
    if is_call:
        grid[:, N] = S_MAX
        grid[0] = np.maximum(S-E, 0) # initial condition
    else:
        grid[:, 0] = E*np.exp(-r*ts)
        grid[:, N] = 0
        grid[0] = np.maximum(E-S, 0)

      # initial conditions for put options


    for m1 in range(1, M+1):
        z_m = np.zeros(N-1)
        z_m[0] = l_0 * grid[m1][0]
        z_m[-1] = u_N * grid[m1][N]
        b_m = grid[m1-1][1:-1] - z_m

        grid[m1][1:-1] = np.linalg.solve(A, b_m)

    return grid
```

## Conclusions

1. During the process of discretization, we can employ various approximations of derivatives to achieve the desired level of numerical precision. 
2. The implicit method, in comparison to the explicit method, exhibits significantly slower computational performance and, moreover, lower accuracy.
3. Careful consideration must be given to the selection of values for N and M due to the stability conditions that need to be satisfied.
